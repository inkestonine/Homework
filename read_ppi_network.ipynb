```python
import pandas as pd
from scipy.sparse import coo_matrix
import networkx as nx
import scipy.sparse as sp
import numpy as np
```

```python
ppi_network = pd.read_csv('data/networks/CPDB_symbols_edgelist.tsv', sep='\t')
```

```python
ppi_network
```

```python
## 这个就是ppi矩阵
ppi_network.head()
```

```python
## gene的名字，后续需要对齐
ppi_network.columns
```

```python
ppi_graph = nx.from_pandas_edgelist(df=ppi_network, source='partner1', target='partner2', edge_attr='confidence')
ppi_network = nx.to_pandas_adjacency(G=ppi_graph)
##get edge
coo_ppi_network = coo_matrix(ppi_network)
edge_index = [coo_ppi_network.row, coo_ppi_network.col]
print('ppi_network have ',len(ppi_network.index), 'genes')
```

```python
data = np.ones(edge_index[1].shape[0])
adj = sp.coo_matrix((data, (edge_index[0], edge_index[1])), shape=(ppi_network.shape[0], ppi_network.shape[1]), dtype=np.float32)
```
